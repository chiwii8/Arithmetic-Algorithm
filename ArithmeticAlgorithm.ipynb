# Example of Arithmetic Algorithm

```python
import sys
import os

arithmetic_path = './src/arith_codding'
sys.path.append(arithmetic_path)

from ArithmeticCode import ArithmeticCodding, EOF
```

```python
print("Example with text")
example_text = "This is a example of arithmetic Algorithm"

example = list(example_text.encode("utf-8"))
example.append(EOF())

print(example_text)
```

```python
codec = ArithmeticCodding()
```

```python
print("Encoding text ...")
bits, table = codec.encode(example)

print("Alphabet used to encode")
print(table.alphabet)
```

```python
print("Decoding text ...")
decoded = codec.decode(bits, table)

result_text = bytes(decoded).decode("utf-8")
print(result_text)
```

```python
print(f"\n{'='*70}")
print("Result:")
print(f"{'='*70}")

print(f"Original text: {example_text}")
print(f"decoded text: {result_text}")
print(f"Match: {'Yes' if result_text==example_text else 'No'}" )

print("repr original:", repr(example_text))
print("repr decoded :", repr(result_text))
print("len original:", len(example_text))
print("len decoded :", len(result_text))
```
